# Alright Folks Lets train our Model

```python
cancer_df
```

```python
# Before we train our model we need to drop target variable from X. Target is the variable we would be predicting
```

```python
X = cancer_df.drop(['target'], axis = 1)
```

```python
X
```

```python
Y = cancer_df['target']
```

```python
Y
```

```python
# Train the model
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.20, random_state = 5)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
Y_train.shape
```

```python
Y_test.shape
```

# its time to have svm in picture

```python
from sklearn.svm import SVC
from sklearn.metrics import classification_report, confusion_matrix
```

```python
svc_model = SVC()
```

```python
svc_model.fit(X_train, Y_train)
```

# Evaluating The Model

```python
y_predict = svc_model.predict(X_test)
```

```python
y_predict
```

```python
cm = confusion_matrix(Y_test, y_predict)
```

```python
sns.heatmap(cm , annot = True)
```

```python
print(classification_report(Y_test, y_predict))
```

# This is the terrible model in the history of SVM Models now we need to improve our Model. 

# woah woah woah lets ask why did our model fail?  because because we didn't normalize the values in dataframes

```python
min_train = X_train.min()
```

```python
min_train
```

```python
range_train = (X_train - min_train).max()
```

```python
range_train
```

```python
x_train_scaled = (X_train - min_train) / range_train
```

```python
x_train_scaled
```

```python
sns.scatterplot(x = X_train['mean area'], y = X_train['mean smoothness'], hue = Y_train)
```

```python
sns.scatterplot(x = x_train_scaled['mean area'], y = x_train_scaled['mean smoothness'], hue = Y_train)
```

```python
min_test = X_test.min()
```

```python
range_test = (X_test - min_test).max()
```

```python
x_test_scaled = (X_test - min_test) / range_test
```

```python
x_test_scaled
```

```python
from sklearn.svm import SVC 
from sklearn.metrics import classification_report, confusion_matrix
svc_model = SVC()
```

```python
svc_model.fit(x_train_scaled, Y_train)
```

```python
y_predict = svc_model.predict(x_test_scaled)
```

```python
y_predict
```

```python
cm = confusion_matrix(Y_test, y_predict)
```

```python
cm
```

```python
sns.heatmap(cm, annot = True, fmt ="d")
```

```python
print(classification_report(Y_test, y_predict))
```

```python
# Thats it 97 percent accuracy. you always need to linearize thedata before training model. Thats what we've learnt
```

```python
# Improve model using Grid Search
```

```python
param_grid = {'C':[0.1, 1, 10, 100], 'gamma' : [1, 0.1, 0.01, 0.001], 'kernel' : ['rbf']}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid = GridSearchCV(SVC(), param_grid, refit = True, verbose = 4)
```

```python
grid.fit(x_train_scaled, Y_train)
```

```python
grid.best_params_
```

```python
grid_predict = grid.predict(x_test_scaled)
```

```python
cm = confusion_matrix(Y_test, grid_predict)
```

```python
sns.heatmap(cm, annot = True)
```

```python
print(classification_report(Y_test, grid_predict))
```
